```python
##Q2.
MongoDB is a popular NoSQL database management system that stores data in a document-oriented format. Here are some of the key features of MongoDB:

Document-oriented: MongoDB stores data in a document-oriented format. Each document is a set of key-value pairs, similar to a JSON object. This makes it easy to store and manipulate complex data structures.

Schema-less: Unlike traditional relational databases, MongoDB does not require a pre-defined schema. This means that you can store data without having to first define the structure, and you can easily add or modify fields as your data evolves.

High Availability: MongoDB supports automatic replication and failover, which makes it highly available and ensures that data is always accessible.

Scalability: MongoDB is designed to scale horizontally by adding more nodes to a cluster. This makes it easy to handle large amounts of data and high traffic volumes.

Indexing: MongoDB supports multiple types of indexing, including primary, secondary, and compound indexes. This allows you to quickly search and retrieve data, even from large data sets.

Aggregation Framework: MongoDB has a powerful aggregation framework that allows you to perform complex queries and transformations on your data. This makes it easy to analyze and visualize data in real-time.

Native JavaScript support: MongoDB supports JavaScript natively, which means you can use JavaScript to interact with the database directly. This makes it easy to integrate with web applications and other JavaScript-based technologies.

Overall, MongoDB's flexible and scalable architecture, combined with its powerful query and aggregation capabilities, make it a popular choice for many modern applications.
```

```python
##Q5.In MongoDB, the find() method is used to query the database and retrieve documents that match a specified set of criteria. The find() method can be used with or without criteria to retrieve documents from a collection.

Here's an example code to demonstrate how to use the find() method in MongoDB:

# Import the necessary libraries
from pymongo import MongoClient

# Connect to the MongoDB client
client = MongoClient()

# Access the database
db = client['my_database']

# Access the collection
collection = db['my_collection']

# Define the query criteria
query = {'name': 'John'}

# Use the find() method to retrieve documents
result = collection.find(query)

# Iterate through the result set and print the documents
for document in result:
    print(document)
In the code above, we first create a MongoDB client and connect to it. We then access the desired database and collection. We define a query criteria that specifies that we want to retrieve all documents where the name field is equal to 'John'. We then use the find() method to retrieve the documents that match the query criteria, and iterate through the result set to print each document.
```

```python
##Q6.
In MongoDB, the sort() method is used to sort the documents in a collection based on one or more fields. The sort() method can be used to sort the documents in ascending or descending order, based on the values of one or more fields.

Here's an example to demonstrate sorting in MongoDB:

Suppose we have a collection named students with documents that look like this:

ObjectId("60b66e0d21f08e5cb7ad4f05"), "name" : "John", "age" : 25, "score" : 80 }
{ "_id" : ObjectId("60b66e0d21f08e5cb7ad4f06"), "name" : "Jane", "age" : 23, "score" : 90 }
{ "_id" : ObjectId("60b66e0d21f08e5cb7ad4f07"), "name" : "Alice", "age" : 22, "score" : 75 }

To sort the documents in the students collection based on the score field in descending order, we can use the sort() method as follows:

# Import the necessary libraries
from pymongo import MongoClient

# Connect to the MongoDB client
client = MongoClient()

# Access the database
db = client['my_database']

# Access the collection
collection = db['students']

# Sort the documents in descending order based on the score field
result = collection.find().sort('score', -1)

# Iterate through the result set and print the documents
for document in result:
    print(document)
In the code above, we first create a MongoDB client and connect to it. We then access the students collection in the my_database database. We use the sort() method to sort the documents in descending order based on the score field, and iterate through the result set to print each document. The -1 parameter passed to the sort() method specifies that we want to sort the documents in descending order. If we wanted to sort the documents in ascending order, we could pass a 1 parameter instead.
```
